```python
#import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import dataset
league_averages = pd.read_csv('/Users/emilysmith/Desktop/ds-studio-ii/leagueaverages.csv')
leagu_averages
```

```python
#clean dataset
league_averages = league_averages.drop(league_averages.index[7:18]) #drop the 7-18 rows since all the values are NaN
#iterates through column names and selects the columns that have 'Unnamed' in the title and drops them
league_averages = league_averages.drop(columns=[col for col in league_averages.columns if 'Unnamed' in col]) #drop all of the columns that have 'Unnamed' in title

league_averages
```

```python
#create a bar graph of 3PAr in WNBA
years = league_averages['Year']
three_par = league_averages['3PAr']
plt.bar(years, three_par, color='orange', edgecolor='black') 
plt.ylim(0.2, 0.4) #sets the limits for y-axis for better understanding of the graph
plt.xlabel('Year')
plt.ylabel('3PAr')
plt.title('Bar Graph of 3PAr in WNBA')
plt.xticks(rotation=45, ha='right') #rotate x-axis labels for better readability
plt.show()
```

```python
#create line chart to show trend of 3PM and 3PA over the years in WNBA
years = league_averages['Year']
three_pm = league_averages['3PM']
three_pa = league_averages['3PA']

plt.plot(years, three_pm, marker='o', label='3PM', color='blue')  # Line for 3PM
plt.plot(years, three_pa, marker='o', label='3PA', color='orange')  # Line for 3PA

# Add labels, title, and legend
plt.xlabel('Years')
plt.ylabel('Amount Per Game')
plt.title('3PM and 3PA Per Game Over Years')
plt.legend()

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Create a line chart to show FG% by distance range over the years in WNBA
years = league_averages['Year']  
distance_ranges = ['2P','0-3', '3-10', '10-16', '16-3P', '3P']  # Distance range columns

# Plot each distance range as a separate line
plt.figure(figsize=(10, 6))
#iterates through the distance range columns and plots them
for col in distance_ranges:
    plt.plot(years, league_averages[col], marker='o', label=col)

# Add labels, title, and legend
plt.xlabel('Year')
plt.ylabel('Attempt Rate or Percentage')
plt.title('Attempt Rate or Percentage by Distance Range Over Time')
plt.legend(title='Distance Range')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability

# Show the plot
plt.tight_layout()
plt.show()
```

```python
#Create pairplot to show the relationships between 3PAr, TS%, eFG%, 3P%, and 3PA in WNBA
#Select the relevant columns
pairplot_data = league_averages[['3PAr', 'TS%', 'eFG%', '3P%', '3PA']]

#plot the pairplot with kde to provide a smoothed version of the distribution of each variable on the diagonal, and set corner=True to only show the lower triangle of the plot
#plot_kws sets the transparency of the points in the scatter plots to 0.7 for better visibility
sns.pairplot(pairplot_data, diag_kind='kde', corner=True, plot_kws={'alpha': 0.7})

# Show the plot
plt.show()
```
